# Using JSON

```python
import json
person = {"name": "Amit", "age": 30, "city": "New York"}
print(type(person))
person_json = json.dumps(person)
print(person_json)
print(type(person_json))
```

```python
# JSON -> dict
person_dict = json.loads(person_json)
print(person_dict)
print(type(person_dict))
```

# Given a list of n integers, find the maximum

```python
l = [1, 5, 2, 7, 8, 10, 4]

mx = l[0]

for i in l:
  if i > mx:
    mx = i

print(mx)
```
